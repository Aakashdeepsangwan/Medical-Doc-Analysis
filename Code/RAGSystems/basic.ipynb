```python
from langchain_community.document_loaders import PyPDFLoader

load_document = PyPDFLoader("Can AI Build Systems (1).pdf").load()
print(type(load_document[0]))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
normal_chunks = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap  = 50).split_documents(load_document)
print(len(normal_chunks))
```

```python

from langchain_huggingface import HuggingFaceEmbeddings
embeddings=  HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2")
```

```python
str_doc  =[doc.page_content for doc in load_document]
str_doc
```

```python
from langchain_experimental.text_splitter import SemanticChunker

Embeddings = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2")
        
        
chunker = SemanticChunker(
            embeddings=Embeddings,
            breakpoint_threshold_type  = "percentile",
            breakpoint_threshold_amount = 95
        )

chunks = chunker.create_documents(str_doc)
print(chunks)
```
